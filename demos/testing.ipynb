# Fetch data from Open Sky

```python
import requests

base_url = "https://opensky-network.org/api"
```

```python
def request_data_in_area(bound_data):
  request_url = f"{base_url}/states/all?lamin={bound_data["lamin"]}&lomin={bound_data["lomin"]}&lamax={bound_data["lamax"]}&lomax={bound_data["lomax"]}"
  
  response = requests.get(request_url)
  
  if response.status_code == 200:
    plane_data = response.json()
    #print(plane_data)

    return plane_data
  else:
    print(f"Failed to retrieve data, {response.status_code}")
```

```python
# Data request for swiss
swiss_data = {
  "lamin": 45.8389,
  "lomin": 5.9962,
  "lamax": 47.8229,
  "lomax": 10.5226
}

swiss_plane = request_data_in_area(swiss_data)
```

# Displaying coordinates on maps

```python
import folium
```

```python
m = folium.Map(
    location=(46.91705697674418, 7.868874418604649),
    control_scale=True,
    zoom_start=7
)
m
```

Adding markers

```python

laavg = 0
loavg = 0
count = 0
for i in swiss_plane_info:
  folium.Marker(
    location=(i[6], i[5]),
    tooltip=i[1],
    popup=f"{i[1]}\n{i[2]}",
    icon=folium.Icon(icon="plane"),
  ).add_to(m)
  
  laavg += i[5]
  loavg += i[6]
  count += 1 
  
  print(f"{i}")

laavg = laavg / count
loavg = loavg / count


```

```python
print(f"laavg : {laavg}, loavg : {loavg}")
```
